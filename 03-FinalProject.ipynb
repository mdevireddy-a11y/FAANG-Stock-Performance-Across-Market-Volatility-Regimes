## Abstract

Financial markets often experience periods of stability and periods of heightened uncertainty, which can significantly affect stock performance. This study investigates how FAANG companies (Facebook/Meta, Apple, Amazon, Netflix, and Google) differ in their return performance and volatility across market regimes defined by the VIX index, a widely used measure of expected market volatility. Specifically, we examine whether some firms exhibit greater resilience during periods of elevated market stress by maintaining more stable returns and lower volatility.

To address this question, we analyze daily stock data for FAANG companies from 2012 to 2019. Daily returns are computed from closing prices, and volatility is measured using the standard deviation of returns as well as a 30-day rolling annualized volatility metric. Market conditions are classified into three regimes based on VIX levels: low volatility (VIX < 20), moderate volatility (20–30), and high volatility (VIX > 30). We compare mean returns, return volatility, and return distributions across these regimes to evaluate how stock behavior changes as market uncertainty increases.

Our results show that all FAANG companies experience positive average returns during low-VIX periods but see significant declines in returns during high-VIX regimes. Volatility increases substantially across all firms as market stress rises, indicating greater risk during uncertain market conditions. However, the magnitude of this increase differs across companies. Netflix consistently exhibits the highest volatility and the largest decline in returns during high-VIX periods, while companies such as Amazon and Google demonstrate relatively smaller increases in volatility, suggesting greater resilience to market stress.

The findings suggest that although FAANG stocks are generally sensitive to market volatility, some firms exhibit more stable performance during turbulent market periods. These results highlight how differences in company characteristics may influence stock resilience during periods of elevated uncertainty.

## Research Question

**How do FAANG companies differ in their daily return performance and volatility across market regimes defined by VIX levels, and which firms exhibit greater resilience during periods of elevated market stress?**

Market conditions are classified into three regimes based on the **CBOE Volatility Index (VIX)**: low volatility (VIX < 20), representing stable market conditions with investor confidence; moderate uncertainty (20 ≤ VIX < 30), indicating elevated investor concern; and high stress (VIX ≥ 30), reflecting acute market turbulence and risk aversion. These thresholds are consistent with established practitioner conventions and prior academic literature (Whaley, 2009).

**Resilience** is defined as a firm’s ability to maintain relatively stable mean daily returns and limit increases in return standard deviation during high-stress regimes (VIX ≥ 30) compared to low-stress conditions (VIX < 20).

## Background and Prior Work

FAANG refers to five major U.S. technology companies: Meta (Facebook), Apple, Amazon, Netflix, and Google (Alphabet). These firms are among the largest publicly traded companies in the world and collectively account for a significant share of the total market capitalization of the S&P 500. Because of their size and market influence, movements in FAANG stock prices can meaningfully affect broader market performance.

Although these firms are frequently grouped together, their underlying business models differ substantially. Apple relies heavily on hardware sales and a premium consumer ecosystem, Google generates most of its revenue through digital advertising, Netflix depends almost entirely on subscription-based streaming services, and Amazon derives significant income from e-commerce and cloud computing through AWS. Prior research comparing large-cap technology firms demonstrates that business model heterogeneity produces meaningfully different return distributions, even among firms that appear similar in market capitalization or sector classification (Hasanhodzic & Lo, 2007). In particular, firms with recurring revenue models such as subscriptions and cloud contracts tend to exhibit lower return volatility than firms reliant on discretionary consumer spending or advertising budgets, which are more sensitive to macroeconomic cycles. This suggests that even within the FAANG group, firms like Netflix and Google — whose revenues depend on subscription retention and advertiser spending respectively — may respond quite differently to periods of elevated market stress.

Prior research also suggests that differences in firm structure and revenue exposure can influence how companies respond to macroeconomic shocks and shifts in investor sentiment more broadly (Fama & French, 1992).
Volatility, defined as the degree to which a stock's price fluctuates over time, is widely used as a measure of financial risk. Research shows that volatility is not random but instead clusters over time and responds strongly to market shocks (Engle & Patton, 2001). During periods of market stress, negative information tends to produce larger increases in volatility than positive news of the same magnitude, a phenomenon known as the leverage effect. These dynamics make volatility a useful metric for evaluating how individual firms react to changing market conditions.

Much of the previous research examines volatility using aggregate measures such as full-sample standard deviation or beta relative to the market. While these methods provide useful information about long-run risk, they often pool together periods of both stable and turbulent market conditions, obscuring important differences in firm behavior across market environments. Existing regime-based volatility studies (e.g., Hamilton, 1989; Ang & Bekaert, 2002) typically focus on index-level behavior or apply hidden Markov models, and do not directly compare individual firm responses within the same sector across fixed, interpretable thresholds. This project addresses that gap by using VIX-based regime classification with pre-specified thresholds to enable direct firm-level comparison within the FAANG group, an approach that is both methodologically transparent and grounded in established practitioner conventions (Whaley, 2009).

### Netflix (NFLX) Daily Stock Prices

This dataset contains daily historical stock market data for Netflix (NFLX), spanning May 2002 through August 2020 and comprising 4,581 trading days. Each row represents one trading day — weekends and market holidays are excluded. All price-related variables are measured in **U.S. dollars (USD) per share**.

- **Open** is the price at which the stock first traded when the market opened that day.
- **High** and **Low** are the highest and lowest prices reached during the trading session, capturing intraday price range and turbulence.
- **Close** is the final traded price at the end of the session.
- **Adj Close** is the closing price adjusted for corporate actions. Netflix underwent a 7-for-1 stock split in 2015, which is why early-era prices appear very small — they have been retroactively scaled down for consistency. **Adj Close is the primary variable used for computing returns.** From it we derive the daily log return (log(Adj Close_t / Adj Close_{t-1})), and volatility is then quantified as the rolling standard deviation of those log returns.
- **Volume** is the number of shares traded during the session (unit: shares). High-volume days often coincide with earnings releases, major news, or broad market stress events.

**Limitations:** The dataset contains no external market variables; VIX data must be merged separately to define volatility regimes. Netflix's business model changed substantially across the dataset's range, from DVD-by-mail (2002–2007) to domestic streaming to global content production, meaning the drivers of volatility differ across eras. The earliest records (2002–2004) reflect a newly public, thinly traded stock with speculative behavior atypical of the mature company. Finally, outliers in financial data typically reflect real events (earnings, crashes, major announcements) rather than errors and must be treated accordingly.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# ── Shared constants ───────────────────────────────────────────────────────────
NUM_COLS = ["Open", "High", "Low", "Close", "Adj Close", "Volume"]

INTERIM_DIR   = Path("data/01-interim")
PROCESSED_DIR = Path("data/02-processed")
INTERIM_DIR.mkdir(parents=True, exist_ok=True)
PROCESSED_DIR.mkdir(parents=True, exist_ok=True)

# ── Netflix ────────────────────────────────────────────────────────────────────
df = pd.read_csv(Path("data/00-raw/Netflix.csv"))
df.columns = [c.strip() for c in df.columns]
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.sort_values("Date").reset_index(drop=True)
for c in NUM_COLS:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# No missing values, no suspicious OHLC rows — confirmed on first run
print(f"Shape: {df.shape}")

# Log returns
df["Daily_Log_Return"] = np.log(df["Adj Close"] / df["Adj Close"].shift(1))
df["Price_Range"]      = df["High"] - df["Low"]

ret = df["Daily_Log_Return"].dropna()
print(f"\nReturn diagnostics — Mean: {ret.mean():.6f}, Std: {ret.std():.6f}, "
      f"Skewness: {ret.skew():.4f}, Excess Kurtosis: {ret.kurtosis():.4f}")

# Outlier flagging (counts only — flagged rows are kept, not removed)
df["Extreme_Return"] = (df["Daily_Log_Return"] <= ret.quantile(0.01)) | (df["Daily_Log_Return"] >= ret.quantile(0.99))
df["High_Volume"]    = df["Volume"] >= df["Volume"].quantile(0.99)
print(f"Extreme return days: {df['Extreme_Return'].sum()} | High-volume days: {df['High_Volume'].sum()}")

print("\nDescriptive statistics:"); display(df[NUM_COLS + ["Daily_Log_Return"]].describe())

fig, axes = plt.subplots(3, 1, figsize=(14, 12))
axes[0].plot(df["Date"], df["Adj Close"], color="royalblue", linewidth=0.8)
axes[0].set_title("Netflix Adjusted Closing Price Over Time"); axes[0].set_ylabel("Adj Close (USD)"); axes[0].grid(alpha=0.3)
axes[1].plot(df["Date"], df["Daily_Log_Return"], color="darkorange", linewidth=0.5, alpha=0.8)
axes[1].axhline(0, color="black", linewidth=0.8, linestyle="--")
axes[1].set_title("Netflix Daily Log Returns"); axes[1].set_ylabel("Log Return"); axes[1].grid(alpha=0.3)
axes[2].plot(df["Date"], df["Daily_Log_Return"].rolling(30).std(), color="firebrick", linewidth=0.8)
axes[2].set_title("Netflix 30-Day Rolling Volatility"); axes[2].set_ylabel("Volatility"); axes[2].grid(alpha=0.3)
plt.tight_layout(); plt.show()

# 0 rows dropped — save processed only
df_clean = df.drop(columns=["Extreme_Return", "High_Volume"], errors="ignore")
df_clean.to_csv(PROCESSED_DIR / "netflix_processed.csv", index=False)
print("Saved."); display(df_clean.head())
```

### VIX (CBOE Volatility Index) Daily Data
This dataset contains daily historical values of the CBOE Volatility Index (VIX), sourced from Yahoo Finance via the yfinance library, spanning January 1990 through June 2024 and comprising 8,685 trading days. Each row represents one trading day — weekends and market holidays are excluded. The VIX is an index value, **not measured in USD per share**, and unlike the equity datasets it contains no Adj Close or Volume columns.

- **Open** is the VIX's first recorded value at the start of the trading day.
- **High** and **Low** are the highest and lowest VIX values observed during the session.
- **Close** is the VIX's final recorded value at the end of the session. **This is the variable used in all regime classifications.** Across the dataset, Close values range from approximately 9.1 to 82.7, with the upper extreme reflecting acute dislocations such as the 2008 financial crisis and the March 2020 COVID-19 shock.

This dataset contains daily VIX prices. We use the `close` column to classify each trading day into one of **three market regimes** consistent with the research question thresholds (Whaley, 2009):

| Regime | VIX threshold | Interpretation |
|---|---|---|
| Low | VIX < 20 | Stable, investor confidence |
| Moderate | 20 ≤ VIX < 30 | Elevated concern |
| High | VIX ≥ 30 | Acute stress / risk aversion |

**Limitations:**  The VIX measures expected S&P 500 volatility rather than the realized volatility of individual FAANG stocks, so it may not fully capture firm-specific risk. It also reflects a 30-day implied horizon and may not represent longer-term structural risk shifts. Regime classifications based on VIX thresholds are best interpreted as snapshots of prevailing market sentiment. Finally, the VIX dataset extends to June 2024, beyond the end dates of the FAANG equity datasets; the analysis will be restricted to the overlapping date range across all datasets.

```python
# ── VIX ────────────────────────────────────────────────────────────────────────
import pandas as pd
from pathlib import Path

INTERIM_DIR   = Path("data/01-interim")
PROCESSED_DIR = Path("data/02-processed")

vix = pd.read_csv("data/00-raw/Vix_daily.csv")
vix.columns = [c.strip().lower() for c in vix.columns]
if "data" in vix.columns:           # handle alternate column name
    vix = vix.rename(columns={"data": "date"})
vix["date"]  = pd.to_datetime(vix["date"], errors="coerce")
vix["close"] = pd.to_numeric(vix["close"], errors="coerce")
vix = vix.sort_values("date").reset_index(drop=True)

print(f"Shape: {vix.shape}")
print(f"Date range: {vix['date'].min().date()} → {vix['date'].max().date()}")
print(f"Missing values:\n{vix.isna().sum()}")

# Three-regime classification — re-derived from raw close
# to guarantee consistency with the research question thresholds
def classify_vix(x):
    if x < 20:   return "Low"
    elif x < 30: return "Moderate"
    else:        return "High"

vix["vix_regime"] = vix["close"].apply(classify_vix)

print("\nVIX regime distribution:")
print(vix["vix_regime"].value_counts())
print("\nVIX close statistics:"); display(vix[["close"]].describe())

vix.to_csv(INTERIM_DIR  / "vix_interim.csv",   index=False)
vix.to_csv(PROCESSED_DIR / "vix_processed.csv", index=False)
print("Saved.")
display(vix.head())
```

### Combining Datasets

All five FAANG stocks are merged into a single long-format dataframe and joined with VIX regime labels. The analysis window is restricted to **2012–2019** to match VIX data availability and avoid partial-year effects at both ends.

**Key design choices:**
- `Daily_Log_Return` (computed in individual cleaning steps above) is used for all return analysis — no switch back to percentage returns.
- 30-day rolling volatility is annualised by multiplying by √252.
- The VIX regime column is named `vix_regime` (Low / Moderate / High) throughout.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

PROCESSED_DIR = Path("data/02-processed")
INTERIM_DIR   = Path("data/01-interim")

# Load individual processed files (output of cleaning cells above)
company_files = {
    "Amazon":   PROCESSED_DIR / "amazon_processed.csv",
    "Apple":    PROCESSED_DIR / "Apple_processed.csv",
    "Facebook": PROCESSED_DIR / "facebook_processed.csv",
    "Google":   PROCESSED_DIR / "google_processed.csv",
    "Netflix":  PROCESSED_DIR / "netflix_processed.csv",
}

dfs = []
for company, path in company_files.items():
    temp = pd.read_csv(path, parse_dates=["Date"])
    temp["Company"] = company
    dfs.append(temp)

faang_df = pd.concat(dfs, ignore_index=True)
faang_df = faang_df[(faang_df["Date"].dt.year >= 2012) &
                    (faang_df["Date"].dt.year <= 2019)].copy()
faang_df = faang_df.sort_values(["Company", "Date"]).reset_index(drop=True)

# Log returns must already exist from the individual cleaning steps.
# If a row is missing Daily_Log_Return (first row per company), it is NaN — acceptable.
# 30-day rolling volatility (annualised) on log returns
faang_df["Volatility"] = (
    faang_df.groupby("Company")["Daily_Log_Return"]
            .transform(lambda x: x.rolling(30).std() * np.sqrt(252))
)

# Merge VIX regimes
vix_df = pd.read_csv(PROCESSED_DIR / "vix_processed.csv", parse_dates=["date"])
faang_df = faang_df.merge(
    vix_df[["date", "vix_regime"]],
    left_on="Date", right_on="date", how="left"
).drop(columns=["date"])

print(f"Combined shape: {faang_df.shape}")
print(f"Date range: {faang_df['Date'].min().date()} → {faang_df['Date'].max().date()}")
print("\nRegime distribution:")
print(faang_df["vix_regime"].value_counts())
print(f"Rows with no VIX match: {faang_df['vix_regime'].isna().sum()}")

# Per-company return diagnostics — documents non-normality for TA feedback
print("\n--- Return distribution diagnostics by company (log returns) ---")
diag = (faang_df.groupby("Company")["Daily_Log_Return"]
               .agg(mean="mean", std="std",
                    skew=lambda x: x.skew(),
                    excess_kurt=lambda x: x.kurtosis())
               .round(6))
display(diag)
print("NOTE: All companies show positive excess kurtosis (heavy tails) and "
      "some skewness. Log returns reduce but do not resolve non-normality. "
      "t-test and OLS results below should be interpreted with this in mind.")

faang_df.to_csv(INTERIM_DIR   / "faang_interim.csv",   index=False)
faang_df.to_csv(PROCESSED_DIR / "faang_processed.csv", index=False)
print("\nSaved combined dataframe.")
display(faang_df.head(10))
```

## Results

### Data Visualization

#### Plot 1: Mean Returns by Regime (Bar Chart)

To evaluate how FAANG companies perform under different market conditions, we compute the mean daily return for each company within VIX-defined market regimes. Daily return is calculated as the percentage change in closing price from one trading day to the next. The mean daily return measures the average profitability of a stock during a given market environment.

This metric is appropriate for our research question because it allows us to compare how stock performance changes during periods of low volatility (VIX < 20), moderate volatility (20–30), and high market stress (VIX > 30). Firms that maintain higher or less negative returns during high-VIX periods can be considered more resilient to market stress.

```python
mean_returns = (
    faang_df
    .groupby(['Company','vix_regime'])['Daily_Log_Return']
    .mean()
    .unstack()
)

mean_returns.plot(kind='bar', figsize=(10,6))
plt.title("Average Daily Returns by VIX Regime")
plt.ylabel("Mean Daily Log Return")
plt.show()
```

The figure shows that all FAANG companies experience positive average daily returns during low VIX periods, indicating that stable market conditions are associated with positive stock performance across the sector. Netflix exhibits the highest mean return during low-volatility markets, followed by Apple and Facebook.

However, during moderate and high VIX regimes, returns decline substantially for all companies. In moderate volatility environments, mean returns become negative for all firms, suggesting that increased uncertainty reduces stock performance.

During high VIX periods, which represent significant market stress, the decline in returns becomes more severe. Netflix experiences the largest drop in mean daily returns, indicating that it is particularly sensitive to market turbulence. In contrast, Amazon shows the smallest decline among the FAANG companies, suggesting relatively stronger resilience during high-volatility conditions.

The results indicate that FAANG companies are highly sensitive to shifts in market volatility. While all firms perform well during stable markets, their returns deteriorate during periods of elevated uncertainty. Among the companies analyzed, Amazon appears to demonstrate the most resilience in high-volatility environments, whereas Netflix appears to be the most vulnerable to market stress.

#### Plot 2: Volatility by Regime

To evaluate how FAANG companies respond to different levels of market stress, we measure return volatility, calculated as the standard deviation of daily returns within each VIX regime. The standard deviation quantifies how much daily returns fluctuate around their average value. Higher values indicate greater variability in returns and therefore higher financial risk.

Volatility is an appropriate metric for addressing our research question because market resilience is not only about maintaining returns but also about limiting large fluctuations in performance during periods of uncertainty. Companies whose volatility increases less dramatically when VIX levels rise can be interpreted as more resilient to market stress.

```python
volatility_by_regime = (
    faang_df
    .groupby(['Company','vix_regime'])['Daily_Log_Return']
    .std()
    .unstack()
)

volatility_by_regime.plot(kind='bar', figsize=(10,6))
plt.title("Volatility of Returns by VIX Regime")
plt.ylabel("Standard Deviation of Returns")
plt.show()
```

The figure shows a consistent pattern across all FAANG companies: volatility increases substantially as VIX levels rise. During low-VIX periods, when market conditions are relatively stable, all companies exhibit lower levels of return variability. However, volatility rises noticeably during moderate VIX conditions and increases even further during high-VIX regimes.

Among the five companies, Netflix exhibits the highest volatility across all regimes, particularly during high-VIX periods, indicating that its returns fluctuate more dramatically during market stress. Apple and Facebook also show relatively large increases in volatility as VIX levels increase.

In contrast, Amazon and Google display comparatively lower volatility levels, even during high-VIX conditions. This suggests that their stock returns are somewhat more stable when market uncertainty rises.

The results indicate that FAANG stocks become significantly more volatile during periods of elevated market stress, as expected given the relationship between the VIX and market uncertainty. However, the magnitude of this increase varies across firms. Amazon and Google appear to demonstrate greater resilience in terms of return stability, while Netflix appears to be the most sensitive to market turbulence.

#### Plot 3: Boxplot of Returns Across Regimes

To further analyze how FAANG companies behave under different market conditions, we examine the distribution of daily returns across VIX regimes using boxplots. A boxplot summarizes several statistical properties of the data, including the median return, interquartile range (IQR), and extreme observations (outliers). This visualization allows us to compare not only the typical return levels but also the spread and variability of returns across market regimes.

This metric is appropriate for our research question because resilience during periods of elevated market stress involves more than just average returns; it also involves how widely returns fluctuate and how frequently extreme positive or negative returns occur. By comparing return distributions across low, moderate, and high VIX regimes, we can observe whether market stress leads to wider return variability and more extreme outcomes.

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=faang_df, x='vix_regime', y='Daily_Log_Return', hue='Company')
plt.title("Distribution of Returns Across VIX Regimes")
plt.show()
```

### TWO-WAY ANOVA Test

To investigate whether FAANG companies differ in their return behavior across varying market conditions, we conducted a two-way ANOVA with Company and VIX_Condition as categorical factors. The model also includes an interaction term to determine whether the effect of market volatility differs across companies. 

The results are evaluated using a signifciance threshold of p = 0.05. A siginificant company effect would indicate that average returns differ across FAANG companies. This suggests that certain firms consistently exhibit higher or lower return performance compared to others, regardless of market conditions.

A significant VIX_Condition effect indicates that overall market volatility influences daily returns. This means that returns tend to change depending on whether the market is in a low, moderate, or elevated volatility regime.

A significant interaction between Company and VIX_Condition indicates that the impact of market volatility on returns differs across companies. This interaction is particularly important for addressing our research question, as it suggests that some companies may maintain more stable returns during periods of elevated market stress, while others experience larger performance shifts.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model_returns = ols('Daily_Log_Return ~ C(Company) * C(vix_regime)', data=faang_df).fit()
anova_returns = sm.stats.anova_lm(model_returns, typ=2)
print(anova_returns)

# ANOVA for volatility
model_volatility = ols('Volatility ~ C(Company) * C(vix_regime)', data=faang_df).fit()
anova_volatility = sm.stats.anova_lm(model_volatility, typ=2)
print(anova_volatility)
```

### T-Test ANALYSIS: FAANG Performance Across Market Regimes

To formally test whether FAANG companies behave differently during periods of market stress versus calm, we run independent-samples t-tests comparing two groups of trading days. We classify days where the VIX condition is **Low** as **Good** market conditions, and days where VIX is **High** as **Bad** market conditions. We run this test twice — once for **daily returns**, and once for **annualized volatility** — to capture both the direction and the risk dimension of performance.

Since we run five simultaneous tests (one per company) for each analysis, we apply a **Bonferroni correction** to control the family-wise Type I error rate, multiplying each raw p-value by 5.


```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

faang_df = pd.read_csv('data/02-processed/faang_processed.csv')
faang_df = faang_df.dropna(subset=['Daily_Log_Return', 'Volatility'])

# Only keep Low (Good) and High (Bad) regime days — Moderate is excluded
# so the comparison is strictly calm vs. stressed markets
faang_df['Market_Regime'] = faang_df['vix_regime'].apply(
    lambda x: 'Good' if x == 'Low' else ('Bad' if x == 'High' else None)
)
faang_df = faang_df.dropna(subset=['Market_Regime'])

companies = sorted(faang_df['Company'].unique())
COLORS = {'Good': '#2ecc71', 'Bad': '#e74c3c'}
COMPANY_COLORS = {
    'Amazon': '#FF9900', 'Apple': '#555555', 'Facebook': '#1877F2',
    'Google': '#4285F4', 'Netflix': '#E50914'
}
good_patch = mpatches.Patch(color=COLORS['Good'], alpha=0.8, label='Good Market (VIX < 20)')
bad_patch  = mpatches.Patch(color=COLORS['Bad'],  alpha=0.8, label='Bad Market (VIX ≥ 30)')

print(f"Good (Low VIX) days : {(faang_df['Market_Regime']=='Good').sum()}")
print(f"Bad  (High VIX) days: {(faang_df['Market_Regime']=='Bad').sum()}")
```

#### Part 1: Daily Returns — Good vs. Bad Market Conditions

We test whether mean daily returns differ significantly between good and bad market regimes for each company. The notched box plots below show the full return distributions — non-overlapping notches indicate statistically significant median differences.

- **H₀**: Mean daily returns are equal across market regimes
- **H₁**: Mean daily returns are lower during bad market conditions

```python
fig, axes = plt.subplots(1, 5, figsize=(18, 6), sharey=True)
fig.suptitle(
    'Daily Returns by Market Regime Across FAANG Companies\n'
    '(Good = Low VIX (<20)  |  Bad = High VIX (≥30))',
    fontsize=14, fontweight='bold', y=1.02)

for ax, comp in zip(axes, companies):
    sub = faang_df[faang_df['Company'] == comp]
    data_good = sub[sub['Market_Regime'] == 'Good']['Daily_Log_Return'] * 100
    data_bad  = sub[sub['Market_Regime'] == 'Bad']['Daily_Log_Return']  * 100
    bp = ax.boxplot([data_good, data_bad],
                    patch_artist=True, notch=True, widths=0.5,
                    medianprops=dict(color='white', linewidth=2),
                    flierprops=dict(marker='o', markersize=2, alpha=0.3))
    bp['boxes'][0].set_facecolor(COLORS['Good']); bp['boxes'][0].set_alpha(0.8)
    bp['boxes'][1].set_facecolor(COLORS['Bad']);  bp['boxes'][1].set_alpha(0.8)
    ax.set_title(comp, fontsize=12, fontweight='bold', color=COMPANY_COLORS[comp])
    ax.set_xticks([1, 2])
    ax.set_xticklabels(['Good\nMarket', 'Bad\nMarket'], fontsize=9)
    ax.axhline(0, color='gray', linestyle='--', linewidth=0.8, alpha=0.6)

axes[0].set_ylabel('Daily Log Return (%)', fontsize=11)
fig.legend(handles=[good_patch, bad_patch], loc='lower center', ncol=2,
           fontsize=10, bbox_to_anchor=(0.5, -0.05), frameon=True)
plt.tight_layout()
plt.show()
```

```python
return_results = []
for comp in companies:
    sub = faang_df[faang_df['Company'] == comp]
    good = sub[sub['Market_Regime'] == 'Good']['Daily_Log_Return'].dropna()
    bad  = sub[sub['Market_Regime'] == 'Bad']['Daily_Log_Return'].dropna()
    t, p = stats.ttest_ind(good, bad, equal_var=False)  # Welch's t-test
    diff = (good.mean() - bad.mean()) * 100
    se = np.sqrt(good.std()**2/len(good) + bad.std()**2/len(bad)) * 100
    ci_lo, ci_hi = diff - 1.96*se, diff + 1.96*se
    p_bonf = min(p * 5, 1.0)
    return_results.append({
        'Company':         comp,
        'Mean Good (%)':   round(good.mean()*100, 3),
        'Mean Bad (%)':    round(bad.mean()*100, 3),
        'Difference (%)':  round(diff, 3),
        '95% CI':          '[{:.3f}, {:.3f}]'.format(ci_lo, ci_hi),
        't-stat':          round(t, 3),
        'p (Bonferroni)':  '<0.001' if p_bonf < 0.001 else round(p_bonf, 4)
    })

return_df = pd.DataFrame(return_results)
display(return_df.style.set_caption('Table 1: T-Test Results — Daily Returns (Good vs. Bad Market)'))
```

```python
fig, ax = plt.subplots(figsize=(11, 5))
bar_colors = [COMPANY_COLORS[c] for c in return_df['Company']]
diffs = return_df['Difference (%)'].values
ci_lo_vals = np.array([float(s.strip('[]').split(',')[0]) for s in return_df['95% CI']])
ci_hi_vals = np.array([float(s.strip('[]').split(',')[1]) for s in return_df['95% CI']])

ax.barh(return_df['Company'], diffs, color=bar_colors, alpha=0.85, height=0.5)
ax.errorbar(diffs, return_df['Company'],
            xerr=[diffs - ci_lo_vals, ci_hi_vals - diffs],
            fmt='none', color='black', capsize=5, linewidth=1.5)
ax.axvline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlabel('Mean Return Difference: Good − Bad Market (%)', fontsize=11)
ax.set_title(
    'T-Test Results: Daily Returns — Good vs. Bad Market Conditions\n'
    '(Error bars = 95% CI  |  Bonferroni correction applied)',
    fontsize=12, fontweight='bold')

for i, row in return_df.iterrows():
    p_val = row['p (Bonferroni)']
    sig = '***' if p_val == '<0.001' else '*'
    p_str = 'p<.001' if p_val == '<0.001' else 'p={}'.format(p_val)
    ax.text(ci_hi_vals[i] + 0.01, row['Company'],
            ' t={}, {} {}'.format(row['t-stat'], p_str, sig),
            va='center', fontsize=9)

ax.set_xlim(left=0, right=ci_hi_vals.max() + 0.25)
plt.tight_layout()
plt.show()
```

##### Returns interpretation
We fail to reject H₀ for all five companies. After Bonferroni correction, no FAANG stock shows a statistically significant difference in mean daily log returns between good (VIX < 20) and bad (VIX ≥ 30) market conditions (all p = 1.0). While the point estimates are in the expected direction, all companies show positive Good − Bad differences, with Netflix showing the largest gap (+2.04%) and Amazon the smallest (+0.82%), the confidence intervals are extremely wide and all cross zero, reflecting the small number of High VIX days in the sample. 

This result does not mean returns are identical across regimes; rather, the High VIX group is too small to detect a difference with adequate statistical power. This is consistent with the limitation noted in the summary: High VIX days are rare by definition, making the Bad group substantially smaller than the Good group and widening confidence intervals considerably. The boxplots support the directional story: Bad market distributions are visibly shifted downward and more spread out for all five companies, most notably Netflix,  but the formal test cannot confirm significance given the sample size constraint.

#### Part 2: Annualized Volatility — Good vs. Bad Market Conditions

We repeat the same t-test framework on annualized rolling volatility (30-day window, scaled by √252). Higher volatility in bad markets is expected, but the **magnitude of the difference** across companies speaks directly to their relative resilience.

- **H₀**: Mean annualized volatility is equal across market regimes
- **H₁**: Mean annualized volatility is higher during bad market conditions

```python
fig, axes = plt.subplots(1, 5, figsize=(18, 5), sharey=True)
fig.suptitle('Annualized Volatility by Market Regime Across FAANG Companies',
             fontsize=14, fontweight='bold')

for ax, comp in zip(axes, companies):
    sub = faang_df[faang_df['Company'] == comp]
    data_good = sub[sub['Market_Regime'] == 'Good']['Volatility'].dropna() * 100
    data_bad  = sub[sub['Market_Regime'] == 'Bad']['Volatility'].dropna()  * 100
    vp = ax.violinplot([data_good, data_bad], positions=[1, 2],
                       showmedians=True, showextrema=False)
    vp['bodies'][0].set_facecolor(COLORS['Good']); vp['bodies'][0].set_alpha(0.75)
    vp['bodies'][1].set_facecolor(COLORS['Bad']);  vp['bodies'][1].set_alpha(0.75)
    vp['cmedians'].set_color('white'); vp['cmedians'].set_linewidth(2)
    ax.set_title(comp, fontsize=12, fontweight='bold', color=COMPANY_COLORS[comp])
    ax.set_xticks([1, 2])
    ax.set_xticklabels(['Good', 'Bad'], fontsize=9)

axes[0].set_ylabel('Annualized Volatility (%)', fontsize=11)
fig.legend(handles=[good_patch, bad_patch], loc='lower center', ncol=2,
           fontsize=10, bbox_to_anchor=(0.5, -0.05), frameon=True)
plt.tight_layout()
plt.show()
```

```python
vol_results = []
for comp in companies:
    sub = faang_df[faang_df['Company'] == comp]
    good = sub[sub['Market_Regime'] == 'Good']['Volatility'].dropna()
    bad  = sub[sub['Market_Regime'] == 'Bad']['Volatility'].dropna()
    t, p = stats.ttest_ind(good, bad, equal_var=False)  # Welch's t-test
    diff = (bad.mean() - good.mean()) * 100
    se = np.sqrt(good.std()**2/len(good) + bad.std()**2/len(bad)) * 100
    ci_lo, ci_hi = diff - 1.96*se, diff + 1.96*se
    p_bonf = min(p * 5, 1.0)
    vol_results.append({
        'Company':        comp,
        'Mean Good (%)':  round(good.mean()*100, 3),
        'Mean Bad (%)':   round(bad.mean()*100, 3),
        'Increase (%)':   round(diff, 3),
        '95% CI':         '[{:.3f}, {:.3f}]'.format(ci_lo, ci_hi),
        't-stat':         round(t, 3),
        'p (Bonferroni)': '<0.001' if p_bonf < 0.001 else round(p_bonf, 4)
    })

vol_df = pd.DataFrame(vol_results)
display(vol_df.style.set_caption('Table 2: T-Test Results — Annualized Volatility (Good vs. Bad Market)'))
```

```python
fig, ax = plt.subplots(figsize=(11, 5))
bar_colors = [COMPANY_COLORS[c] for c in vol_df['Company']]
diffs = vol_df['Increase (%)'].values
ci_lo_vals = np.array([float(s.strip('[]').split(',')[0]) for s in vol_df['95% CI']])
ci_hi_vals = np.array([float(s.strip('[]').split(',')[1]) for s in vol_df['95% CI']])

ax.barh(vol_df['Company'], diffs, color=bar_colors, alpha=0.85, height=0.5)
ax.errorbar(diffs, vol_df['Company'],
            xerr=[diffs - ci_lo_vals, ci_hi_vals - diffs],
            fmt='none', color='black', capsize=5, linewidth=1.5)
ax.axvline(0, color='gray', linestyle='--', linewidth=1)
ax.set_xlabel('Volatility Increase in Bad vs. Good Markets (%)', fontsize=11)
ax.set_title(
    'T-Test Results: Annualized Volatility — Good vs. Bad Market Conditions\n'
    '(Error bars = 95% CI  |  Bonferroni correction applied)',
    fontsize=12, fontweight='bold')

for i, row in vol_df.iterrows():
    p_val = row['p (Bonferroni)']
    sig = '***' if p_val == '<0.001' else '*'
    p_str = 'p<.001' if p_val == '<0.001' else 'p={}'.format(p_val)
    ax.text(ci_hi_vals[i] + 0.1, row['Company'],
            ' t={}, {} {}'.format(row['t-stat'], p_str, sig),
            va='center', fontsize=9)

ax.set_xlim(left=0, right=ci_hi_vals.max() + 3)
plt.tight_layout()
plt.show()
```

##### Volatility interpretation

We **reject H₀** for four of the five companies. Volatility is significantly higher during bad market conditions for Amazon (p = 0.008), Apple (p = 0.006), Google (p = 0.007), and Netflix (p = 0.023) after Bonferroni correction. Facebook fails to reach significance (p = 0.058), though the point estimate (+6.55%) is in the expected direction.

In terms of magnitude, **Google** shows the largest absolute volatility spike (+19.0 percentage points), followed by Apple (+16.5%), Netflix (+16.5%), Amazon (+11.7%), and Facebook (+6.6%). However, Google's large increase should be interpreted alongside the fact that it has the **lowest baseline volatility in good markets (20.8%)** — nearly half that of Netflix (42.7%) — meaning it starts from a much lower floor even though its stress-period jump is the largest in absolute terms.

**Netflix** has by far the highest volatility in both regimes (42.7% in good markets, 59.2% in bad), confirming it as the most volatile FAANG stock regardless of market conditions. The violin plots reinforce this — Netflix's good-market distribution is already wide and right-skewed, while its bad-market distribution shifts substantially upward.

Facebook's non-significant result (p = 0.058, just above the Bonferroni-corrected threshold) combined with the smallest increase (+6.6%) suggests it is comparatively less sensitive to market stress in terms of volatility, though the limited number of High VIX days reduces statistical power for all companies and this result should be interpreted cautiously.

#### Summary

Both t-tests converge on the same conclusion: **market conditions as measured by VIX regime have a significant and consistent effect on FAANG stock behavior.**

| Finding | Detail |
|---|---|
| **Returns in bad markets** | Returns are lower during High VIX periods for all companies but not statistically significant after Bonferroni correction (all p = 1.0), likely due to limited High VIX observations |
| **Volatility rises in bad markets** | Significantly higher for Amazon, Apple, Google, Netflix (all p < 0.05); Facebook borderline (p = 0.058) |
| **Most resilient** | Google — lowest baseline volatility (20.8%); Apple — smallest return gap |
| **Least resilient** | Netflix — highest volatility in both regimes (42.7% good, 59.2% bad) |


> **Limitation:** Bad market days (n=994) are far fewer than good market days (n=8,821), which widens confidence intervals for bad-market estimates. Future work could address this through bootstrapped confidence intervals or a matched-sample design to equalize group sizes.

To address this, we defined three market regimes based on VIX levels (low: VIX < 20, moderate: 20 ≤ VIX < 30, high: VIX ≥ 30) and conducted a two-way ANOVA to examine differences in daily returns and volatility across companies and market conditions. Overall, our findings indicate that FAANG companies are more similar in return behavior but significantly different in risk.

In contrast, the volatility analysis revealed substantial differences across firms. Both the company effect and the interaction between company and market regime were highly significant (p < 0.001), indicating that while returns may behave similarly, the magnitude of risk varies meaningfully across companies and responds differently to market stress. This distinction is important, as two firms may exhibit similar average returns while experiencing vastly different levels of fluctuation. These findings suggest that firm-specific characteristics—such as revenue structure, predictability, and investor sentiment—play a critical role in shaping volatility, even when overall return trends are aligned. Taken together, these results extend prior literature in a practical way.
